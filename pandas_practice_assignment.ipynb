```python
chip_data["item_price"] = chip_data["item_price"].apply(lambda x:float(x[1:-1]))
chip_data.head()
```

```python
# how much was the revenue for the period of dataset.
# revenue = quanitity * item_price
revenue = (chip_data["quantity"] * chip_data["item_price"]).sum()
print("revenue is" + " " + "$" + str(revenue))
```

```python
#print a dataset with only 2 columns item_name and item_price
item_data = chip_data[["item_name","item_price"]]
item_data.head()
```

```python
# remove the duplicates from the item name and price
filtered = chip_data.drop_duplicates(["item_name","item_price"])

# on;y the products with quantity = 1
one_prod = filtered[filtered["quantity"] == 1]

# select only item name and price
selected = one_prod[["item_name","item_price"]]

# sort in decending order of price
selected.sort_values(by="item_price",ascending=False)
```

```python
#what was the quanitity of most expensive item ordered
# first find the most expensive item.
chip_data.sort_values(by="item_price",ascending=False)
```
